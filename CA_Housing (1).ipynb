```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing(as_frame=True)
df = housing.frame
```

```python
df.head()
```

```python
df.describe().T
```

```python
df.hist(
    bins = 50,
    figsize = (16,10))
plt.show
```

```python
sns.set_theme(style="ticks")
sns.pairplot(df)
```

```python
sns.scatterplot(df,
    x="MedInc",
    y="MedHouseVal",
    hue="HouseAge",
    size="AveRooms",
    palette="viridis")
plt.show()
```

```python
sns.scatterplot(df,
    x="Longitude",
    y="Latitude",
    hue="MedHouseVal",
    size="Population",
    palette="magma")
plt.show()
```
